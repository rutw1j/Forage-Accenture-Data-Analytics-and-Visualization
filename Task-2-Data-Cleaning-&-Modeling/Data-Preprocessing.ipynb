# Data Preprocessing on Skill Buzz's Social Media Datasets

## Importing Required Libraries and Loading Datasets

```python
# Importing Libraries
import numpy as np
import pandas as pd

# Loading Datasets
content = pd.read_csv("https://raw.githubusercontent.com/rutw1j/Forage-Accenture-Data-Analytics-and-Visualization/main/Resources/Content.csv")
reactions = pd.read_csv("https://raw.githubusercontent.com/rutw1j/Forage-Accenture-Data-Analytics-and-Visualization/main/Resources/Reactions.csv")
reaction_types = pd.read_csv("https://raw.githubusercontent.com/rutw1j/Forage-Accenture-Data-Analytics-and-Visualization/main/Resources/Reaction-Types.csv")
```

## Dropping unrequired columns and Null Values

```python
# Dropping columns from content-dataset
content.drop(["Unnamed: 0", "User ID", "URL"], axis=1, inplace=True)

# Dropping columns from reactions-dataset
reactions.drop(["Unnamed: 0", "User ID"], axis=1, inplace=True)

# Dropping columns from reaction-types-dataset
reaction_types.drop("Unnamed: 0", axis=1, inplace=True)

# Removing null values from reactions-dataset
reactions.dropna(inplace=True)
```

## Transforming Category Columns and Renaming Columns

```python
# Lower case all values in category column
content['Category'] = content['Category'].str.lower()

# removing double quote (") from category column
content['Category'] = content['Category'].str.replace('"',  '', regex=True)

# Renaming columns from respective datasets
content.rename(columns={"Content ID":"Content-ID", "Type":"Content-Type"}, inplace=True);
reactions.rename(columns={"Content ID":"Content-ID", "Type":"Reaction-Type"}, inplace=True);
reaction_types.rename(columns={"Type":"Reaction-Type"}, inplace=True);
```

## Merging the Datasets and Rearranging Column Sequence

```python
# Merging Reactions-data and Content-Data
merged_data = pd.merge(reactions, content, on='Content-ID', how="left")

# Merging the newly Merged-data and Reactions-Data
merged_data = pd.merge(merged_data, reaction_types, on="Reaction-Type", how="left")

# Rearranging sequence of columns
merged_data = merged_data[['Content-ID', 'Datetime', 'Content-Type', 'Category', 'Reaction-Type', 'Sentiment', 'Score']]
merged_data.sample(20)
```

## Grouping data by Category

```python
# Grouping data by categories and aggregating using sum of scores
top_categories = merged_data.groupby('Category').agg({"Reaction-Type":"count", "Score":"sum"}).reset_index().sort_values(by="Score", ascending=False)
top_categories.head(15)
```

## Exporting dataframe to CSV file

```python
#merged_data.to_csv("/content/Merged-Data.csv", index=False)
#top_categories.to_csv("/content/Top-Categories.csv", index=False)

with pd.ExcelWriter('/content/Merged-Data-with-Categories.xlsx') as writer:
    merged_data.to_excel(writer, sheet_name='Cleaned-Data', index=False)
    top_categories.to_excel(writer, sheet_name='Top-Categories', index=False)
```
